```python
from sage.all import *
x, y = var('x, y')
```

```python
g1 = (x^3 - x^2 - sqrt(x^4 - 4*x^3 + 4*x^2 + 4) - 2)/x^5
```

```python
g2 = (x^3 - x^2 + sqrt(x^4 - 4*x^3 + 4*x^2 + 4) - 2)/x^5
```

```python
def h1(a, b):
    return max(- log(abs(a+b*1j)) - log(abs(g1(x=a+b*1j))), -3)
def h2(a, b):
    return max(- log(abs(a+b*1j)) - log(abs(g2(x=a+b*1j))), -3)
```

```python
redim = (-4, 4)
imdim = (-4, 4)
surface_reso = 100
p = plot3d(h1, redim, imdim, plot_points=surface_reso, color='purple', opacity=0.5)
p += plot3d(h2, redim, imdim, plot_points=surface_reso, color='green', opacity=0.5)

# degeneracy 4 critical point
p += point3d((2, 1/8, h2(2, 0)), color='red', size=50)

# degeneracy 2 points
# p += point3d((3.236067977499790, g2(x=3.236067977499790), h2(3.236067977499790, 0)), color='red', size=50)
# p += point3d((-1.236067977499790, g2(x=-1.236067977499790), h2(-1.236067977499790, 0)), color='red', size=50)
p.show()
```
